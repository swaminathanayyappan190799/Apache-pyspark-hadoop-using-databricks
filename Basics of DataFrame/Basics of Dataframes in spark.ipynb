```
#Importing spark
import pyspark
```

```
#Reading the input data in the form of table.
data=spark.read.table('people')
```

```
#Prints the entire input data.
data.show()
```

```
#Gives the schema of the input data.
data.printSchema()
```

```
#gives the statistical terms of the input data.
data.describe().show()
```

```
#Retrieves first two rows of the input data.
data.head(2)
```

```
#Retrieves the last row of the input data.
data.tail(1)
```

```
#Selecting age column from the input data.
data.select('age').show()
```

```
#retrieving multiple columns from the input data.
data.select(['age','name']).show()
```

```
#Retrieving the age of the second person in the input data.
data.head(2)[1][0]
```

```
#Retrieving the name from the last row.
data.tail(1)[0][1]
```

```
#adds new column to the input data.
data.withColumn('new age',data['age']).show()
```

```
#manipulating input data
data.withColumn('new double age',data['age']*2).show()
```

```
#Renaming a column on the input data.
data.withColumnRenamed('age','altered_age').show()
```

```
#To perform SQL activities on the input data we want to create a view for an input data.
data.createOrReplaceTempView('peoplesql')
```

```
#retrieving entire table 
spark.sql('SELECT * FROM peoplesql').show()
```

```
#Retrieving names who has age grater than 15.
spark.sql('Select name from peoplesql where age>15').show()
```
