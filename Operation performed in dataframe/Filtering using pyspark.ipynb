```
#Importing pyspark.
import pyspark
```

```
#Reading the input data.
data=spark.read.table('appl_stock')
```

```
#Viewing the input data.
data.show()
```

```
#Iferring schema of the input data.
data.printSchema()
```

```
#Filtering input data with a condition.
data.filter('High > 200').show()
```

```
#Filtering input data with a condition by selecting two columns for displaying as output. 
data.filter('open<100').select(['Date','Volume']).show()
```

```
#Filtering input data based on multiple conditions.
data.filter((data['Low']>200)&(data['High']>100)).select(['Volume','Adj Close']).show()
```

```
#Filtering input data based on a condition by selecting only one column for displaying it as output.
data.filter(data['Low']>300).select('High').show()
```

```
#Filtering the data and converts the result into an list using .collect() 
result=data.filter('High >300').collect()
```

```
#Displaying the result.
result
```

```
#Retrieving the first value in the result list. 
result[0]
```

```
#Retrieving the open value for the first value in the result list.
result[0][1]
```

```
#Converts the first row of the result list into a dictionary and assigning it to the row_value variable.
row_value=result[0].asDict()
```

```
#printing the row_value dictionary.
row_value
```

```
#printing the low value from the row_value dictionary.
row_value['Low']
```
