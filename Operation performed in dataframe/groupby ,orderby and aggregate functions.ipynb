```
#Importing Pyspark
import pyspark
```

```
#Reading input data using spark.
data=spark.read.table('sales_info')
```

```
#Inferring Schema of the input data.
data.printSchema()
```

```
#Viewing the input data.
data.show()
```

```
#Groupby functions are used to compress rows based on the column selected.
#Grouping company names with their average value of sales.
data.groupby("Company").mean().show()
```

```
#Grouping company names with their sum value of sales.
data.groupby("Company").sum().show()
```

```
#Grouping company names with their count.
data.groupby("Company").count().show()
```

```
#aggregating sales column with their minimum value.
data.agg({'Sales':'min'}).show()
```

```
#Groupby the company name and aggregating it by the max of sales produced by each company.
data.groupby("company").agg({'sales':'max'}).show()
```

```
#Importing Functions from pyspark.
from pyspark.sql.functions import format_number,avg,stddev
```

```
#Selecting average of sales and renaming the column as average of sales.
data.select(avg("Sales").alias("Average of sales")).show()
```

```
#Selecting the standard deviation of sales column and formating the result with a precision of 2.
data.select(stddev('Sales').alias('stddev of sales')).select(format_number('stddev of sales',2).alias('stddev of sales')).show()
```

```
#Orderby helps to order the rows in ascending / descending order.
#Ordering the sales column based on ascending order.
data.orderBy('Sales').show()
```

```
#Ordering the company name in ascending order.
data.orderBy('Company').show()
```

```
#Ordering the company name in descending order
data.orderBy(data['Company'].desc()).show()
```
